## The break statements

### break allows programmers to terminate the loop, break encountered inside a loop, the loop is immediately terminated.

```python
for i in range(100):
    if i == 14:
        break
    else:
        print(i, end=' ')
```

```python
count=35
for x in range(10):
    count =count-1
    if x==2:
        break
print(count)
```

```python
z=1
while z<5:
    if z%7==0:
        break
    z+=2
print(z)
```

## The continue statements

### skips the current iterations and also skip the remaining statements within the body

```python
for i in range(20):
    if i == 7:
        continue
    else:
        print(i, end=' ')
```

```python
for i in range(70):
    if i%7 == 0 or i%9 == 0:
        continue
    else:
        print(i, end=' ')
```

```python
string = input('Enter any line: ')

print('line without spaces: ')
for i in string:
    
    if i == ' ':
        continue
    else:
        print(i, end='')
```

```python
My_str ='I LOVE PYTHON'
count=0
for char in My_str:
    if char=='O':
        continue
    else:
        count+=1
print(count)
```

```python
My_str ='I LOVE PYTHON'
count=0
for char in My_str:
    if char=='E':
        break
    else:
        count+=1
print(count)
```
