### Assigning value to attribute

```python
class Rectangle:  # Defining the class
    length = 0   # Class attribute
    breadth = 0  # Class attribute
R1 = Rectangle()  # class instance
print('Lenght of rectangle:',R1.length)   # accessing the attributes value
print('Breadth of the rectangle:',R1.breadth)
print('Are of the rectangle:',R1.length*R1.breadth)

R1.length = 20   # Assigning value to Attribute
R1.breadth = 30
print('Lenght of rectangle:',R1.length)
print('Breadth of the rectangle:',R1.breadth)
print('Are of the rectangle:',R1.length*R1.breadth)
```

## Self-parameter and adding methods to class

### The self-parameter is used in the implimentation of the method, Self-parameter reference the object itself

#class Class_name
     def method_name(self, parameter list)

```python
class methodDemo:
    def display_msg(self):
        print('Welcome to python programming!')
obj = methodDemo()
obj.display_msg()
```

```python
class adding:
    def addition(self, a, b):
        return a+b
    
obj=adding()
obj.addition(1, 3)
```

```python
class area_of_circle:
    def circle(self, r):
        return 3.14*r**2
    
obj1=area_of_circle()
print('Area of the circle',obj1.circle(5))
```

```python
class rect_area:
    def rectangle(self, l, b):
        return l*b
    
obj2=rect_area()
print('Area of the rectangle',obj2.rectangle(5, 4))
```

### self parameter with method
### calling method inside method

```python
class self_demo:
    def method_A(self):
        print('Call from A')
        print('Wow, calling A')
        
    def method_B(self):
        print('Call from B')
        self.method_A()
        
obj=self_demo()
obj.method_B()
```

## Accessibility

### Public, private and protected all are public by default but you can make it private(hide) by adding __ to methods name

```python
class BankDetails:
    def __init__(self):
        self.Name = 'Naveen'
        self.__BankAccountnum = '10110'
    def Display(self):
        print('Name: ', self.Name)
        print('Account Num: ',self.__BankAccountnum)
        
D1 = BankDetails()
D1.Display()
```

## __init__ Method (Constructor or initialiser)

```python
class circleArea:
    def area(self, pi, r):
        return pi*r**2
    
D1=circleArea()
D1.area(3.14, 5)
```

```python
class circle:
    def __init__(self, pi):
        self.pi = pi
    
    def circlearea(self, r):
        return self.pi*r**2
    
D1 =circle(3.142)
print('Area of the circle is',D1.circlearea(5))
```

```python
class circle:
    r = 0
    pi = 0
    def __init__(self):
        self.r=5
        self.pi=3.14
    def circlecal(self):
        print('Radius of the circle: ',self.r)
        return self.pi*self.r**2
d1=circle()
print('Area of the circle is',d1.circlecal())
```

## Inheritance:
## Creation of a new subclass using existing class with no modification

```python
class A:
    print('Hello I am Class A')
    
class B(A):
    print('It\'s me Class B')
          
obj = B
obj
```

```python
class point:    # Base or super class
    def coordinates(self, x, y):
        self.x=x
        self.y=y
        
class new_points(point):     # Subclass or Derived class
    def new_draw(self):
        print('Locate point X',self.x,'on X axis')
        print('Locate point Y',self.y,'on Y axis')   
        
p=new_points()
p.coordinates(10, 20)
p.new_draw
```
