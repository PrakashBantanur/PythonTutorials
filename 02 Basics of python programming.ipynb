## 2.7 MULTIPLE ASSIGNMENTS

### Var1, Var 2, Var3 .........VarN = Exp1, Exp2, Exp3.........ExpN

### Swapping using multiple assignment

```python
p = 10
q = 20

print('p = ',p)
print('q = ',q)

p, q = q, p

print('\nSwapped results')
print('p = ',p)
print('q = ',q)
```

### Note: programmer doesn't know what value he can enter, they may enter int, float, string, complex by using eval() python auto determines the type

## 2.11 FORMATTING NUMBER AND STRINGS
### format(item, format-specifier)

```python
x = 1.6543
print(x)
format(x, ".2f")
```

```python
pi = 3.142
radius = eval(input('Enter the radius: '))
area = 2 * pi * radius

print('Area of the circle: ', area)
print('Area of the circle with format: ', format(area, ".2f"))
```
